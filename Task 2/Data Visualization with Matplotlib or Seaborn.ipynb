# Data Visualization with Matplotlib/Seaborn

```python
# importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading dataset
titanic = pd.read_csv('Dataset/Titanic-Dataset.csv')
titanic.head()
```

### Create a histogram to visualize the distribution of a numerical variable.

```python
sns.histplot(data=titanic, x='Age', bins=40, color='blue')
plt.title('Age Distribution')
plt.show()
```

### Plot a bar chart to show the frequency of categories in a categorical variable.

```python
sns.barplot(data=titanic, x='Sex', y="Survived", palette='Blues')
plt.title('Survival Rate by Gender')
plt.show()
```

### Create a scatter plot to visualize the relationship between two numerical variables.

```python
sns.scatterplot(data=titanic, x='Age', y="Fare", color='blue')
plt.show()
```
